## Data preprocessing for D3 visualization

### This notebook is used to preprocess the data used for D3 visualization library on Observable to explain the popularity of Kpop group

```python
import pandas as pd
```

```python
df = pd.read_csv('merged_kpop_song_features.csv')
```

```python
df.head(10)
```

```python
df = df.drop(columns=['Unnamed: 0', 'key', 'valence', 'tempo', 'duration_ms', 'time_signature', 'release_date', 'song_name','song_name_formatted'])
```

```python
df =df.groupby(['artist_name']).sum()
```

```python
df
```

```python
df.to_csv('processeddata.csv')
```
